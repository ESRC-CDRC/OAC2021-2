```python
import pyarrow.parquet as pq
import pandas as pd
from clustergram import Clustergram
import sklearn
#import cudf
import matplotlib.pyplot as plt

import pickle
```

```python
dat = pd.read_csv('/Users/jakubwyszomierski/Desktop/PhD/GIT/OAC2021/Data/Clean/Final_variables/Aged_Scotland.csv')
dat = dat.set_index('Geography_Code')
dat = dat.drop('country', axis=1)
```

```python
### Supergroups clustering

cgram = Clustergram(range(1, 12), backend='sklearn', n_init=200)
cgram.fit(dat)
cgram.plot(figsize=(12, 8))
plt.savefig('/Users/jakubwyszomierski/Desktop/PhD/GIT/OAC2021/Plots/Clustergrams/Supergroups_OAC.pdf')
```

```python
#Groups clustering


dat = pd.read_csv('/Users/jakubwyszomierski/Desktop/PhD/GIT/OAC2021/Data/Clustering/OAC_assignment.csv')
dat = dat.set_index('Geography_Code')
dat = dat.drop(['Group','Subgroup'], axis=1)
dat
### Supergroups clustering

columns = ['Supergroup8']

for column in columns:
    print(column)
    for i in dat[column].unique():
        print (i)
        filtered_df = dat[dat[column] == i]
        filtered_df = filtered_df.drop('Supergroup8', axis=1)
        cgram = Clustergram(range(1, 6), backend='sklearn', n_init=200)
        cgram.fit(filtered_df)
        cgram.plot(figsize=(12, 8))
        filename = '/Users/jakubwyszomierski/Desktop/PhD/GIT/OAC2021/Plots/Clustergrams/Groups/Supergroup_' + str(i) + '.pdf'
        plt.savefig(filename)
```
